# PostgreSQL
- Invented in Berkeley Computer Science Department
- Started in 1986, to create a database system with minimal features needed to support multiple data types

## Installation
- Visit https://www.enterprisedb.com/downloads/postgres-postgresql-downloads
- Download most recent version for your system
- Install, specifying the location directory
- Select Components: PostgreSQL Server, pgAdmin 4, Command Line Tools
- The defaults for the following should be fine:
- Select Storage Directory
- Select Password
- Select Port
- Select Locale

## Connect to a Database
There are two ways to connect to the postgreSQL database:
- SQL Shell (psql)
- pgAdmin 4 

### SQL Shell (psql)
1) Open SQL Shell (psql) programme
2) Enter name of server ("localhost" is a good default)
3) Select Database = "postgres"
4) Select Port
5) Enter Username: suggestion is "postgres"
6) Enter Password - the password chosen at installation
7) Should see a version like psql (15.2) and the `postgres=#` 
- From here, we can write SQL queries e.g. `SELECT version();`

### pgAdmin4
1) Open pgAdmin4 programme
2) Choose a master password
3) Find databases in the left column
4) Right-click on your database and choose `Query Tool`
5) We can write SQL queries and execute them here

## Database Commands

### Create a Table

```python
-- This is a comment in SQL
CREATE TABLE table_name(
    column1 VARCHAR(255), -- We have to specify datatype
    column2 INT,
    column3 datatype,
);
```

```python
SELECT * FROM table_name;
```

### Insert Data

```python
INSERT INTO table_name(column1, column2, column3)
VALUES 
    ('value1', 123, 'value3'),
    ('value4', 456, 'value6');

-- SQL Shell will return: INSERT 0 2 meaning it has inserted 2 rows
```

### Fetch Data

```python
SELECT * FROM table_name;

SELECT column1, column2 FROM table_name;
```

### Alter Data

```python
-- Add a new column to the table called column4 of type DATE:
ALTER TABLE table_name
ADD column4 DATE;
```

```python
-- Update a specific entry with the UPDATE statement:
UPDATE table_name
SET column4 = '2024-01-01'
WHERE column1 = 'value1';

-- The WHERE is very important to avoid updating all rows in the table!
```

```python
-- Use ALTER COLUMN to change datatype of a column
ALTER TABLE table_name
ALTER COLUMN column2 TYPE VARCHAR(30);
```

```python
-- DROP COLUMN statement to remove a column
ALTER TABLE table_name
DROP COLUMN column3;

-- DELETE statement to remove rows from a table
DELETE FROM table_name
WHERE column1 = 'value4';
-- WHERE clause is very important to avoid deleting all rows in the table!

-- TRUNCATE TABLE to remove all rows from a table
TRUNCATE TABLE table_name;
```

```python
-- To drop an existing table use DROP TABLE statement
DROP TABLE table_name;
-- table_name no longer exists 
```

## PostgreSQL Syntax

### Operators
- `=`	Equal to
- `<`	Less than
- `>`	Greater than
- `<=`	Less than or equal to
- `>=`	Greater than or equal to
- `<>`	Not equal to
- `!=`	Not equal to
- `LIKE`	Check if a value matches a pattern (case sensitive)
    - Patterns can use wildcards `%` and `_` e.g.
        - `SELECT * FROM cars`
        - `WHERE model LIKE V%` would return Volvo and Volkswagen 
        - `%` means any number of characters, `_` means a single character
- `ILIKE`	Check if a value matches a pattern (case insensitive)
- `AND`	Logical AND
- `OR`	Logical OR
- `IN`	Check if a value matches any value within a provided list
- `BETWEEN`	Check if a value is within a specified range (inclusive)
- `IS NULL`	Check if a value is NULL
- `NOT`	Makes a negative result e.g. NOT LIKE, NOT IN, NOT BETWEEN


### JOINS

Used to combine rows from two or more tables.
- `INNER JOIN`
- `LEFT JOIN`
- `RIGHT JOIN`
- `FULL JOIN`
- `CROSS JOIN`
