# PostgreSQL
- Invented in Berkeley Computaer Science Department
- Started in 1986, to create a database system with minimal features needed to support multiple data types

## Installation
- Visit https://www.enterprisedb.com/downloads/postgres-postgresql-downloads
- Download most recent version for your system
- Install, specifying the location directory
- Select Components: PostgreSQL Server, pgAdmin 4, Command Line Tools
- The defaults for the following should be fine:
- Select Storage Directory
- Select Password
- Select Port
- Select Locale

## Connect to a Database
There are two ways to connect to the postgreSQL database:
- SQL Shell (psql)
- pgAdmin 4 

### SQL Shell (psql)
1) Open SQL Shell (psql) programme
2) Enter name of server ("localhost" is a good default)
3) Select Database = "postgres"
4) Select Port
5) Enter Username: suggestion is "postgres"
6) Enter Password - the password chosen at installation
7) Should see a version like psql (15.2) and the `postgres=#` 
- From here, we can write SQL queries e.g. `SELECT version();`

### pgAdmin4
1) Open pgAdmin4 programme
2) Choose a master password
3) Find databases in the left column
4) Right-click on your database and choose `Query Tool`
5) We can write SQL queries and execute them here

### Create a Table
We have to specify the datatype of each column e.g. `VARCHAR(255)`

```python
CREATE TABLE table_name(
    column1 VARCHAR(255),
    column2 INT,
    column3 datatype,
);
```

```python
SELECT * FROM table_name;
```

### Insert Data

```python
INSERT INTO table_name(column1, column2, column3)
VALUES 
    ('value1', 123, 'value3'),
    ('value4', 456, 'value6');
```
